```python
import pandas as pd
import numpy as np 
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
dist=load_iris()
dist
```

```python
df=pd.DataFrame(dist['data'],columns=dist['feature_names'])
df
```

```python
y=pd.DataFrame(dist['target'],columns=["target"])
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
scaler.fit(df)
scaled_features=scaler.transform(df)
```

```python
df_feat=pd.DataFrame(scaled_features,columns=dist["feature_names"])
```

```python
from sklearn.cross_validation import train_test_split
```

```python
 X_train,X_test,y_train,y_test=train_test_split(df_feat,y,test_size=0.3,random_state=101)
```

```python
knn=KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train,y_train)
```

```python
predictions=knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
knn.score(X_test,y_test)
```
